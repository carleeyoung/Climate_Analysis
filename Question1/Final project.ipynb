```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make a reference to the Unemployment.csv file path
# File to Load 
#file_to_load = "Resources/Unemployment.csv"
unemployment_path = "Unemployment.csv"

# Import the Unemployment.csv file as a DataFrame
df = pd.read_csv(unemployment_path, encoding="utf-8")

df.head(1)
```

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame
#only keep; 'Stabr', 'area_name',
       #' Employed_2014 ', ' Unemployed_2014 ', 'Unemployment_rate_2014', 
       #' Employed_2015 ', ' Unemployed_2015 ', 'Unemployment_rate_2015',
       #' Employed_2016 ', ' Unemployed_2016 ', 'Unemployment_rate_2016',
       #'Employed_2017','Unemployed_2017', 'Unemployment_rate_2017',
       #'Employed_2018', 'Unemployed_2018', 'Unemployment_rate_2018', 
       #' Employed_2019 ', ' Unemployed_2019 ', 'Unemployment_rate_2019',
       #'Median_Household_Income_2018',
reduced_df = df[["Stabr", "area_name", 
                 ' Employed_2014 ', ' Unemployed_2014 ', 'Unemployment_rate_2014',
                 ' Employed_2015 ', ' Unemployed_2015 ', 'Unemployment_rate_2015',
                 ' Employed_2016 ', ' Unemployed_2016 ', 'Unemployment_rate_2016',
                 'Employed_2017','Unemployed_2017', 'Unemployment_rate_2017',
                 'Employed_2018', 'Unemployed_2018', 'Unemployment_rate_2018',
                 ' Employed_2019 ', ' Unemployed_2019 ', 'Unemployment_rate_2019',
                 'Median_Household_Income_2018'
                ]]
                 

reduced_df.head()
```

```python
# Rename columns so that they are differentiated
reduced_df = reduced_df.rename(columns={"Stabr": "stabr",
                                        "area_name":"AreaName",
                 ' Employed_2014 ':"Employed2014", 
                 ' Unemployed_2014 ':"Unemployed2014", 
                 'Unemployment_rate_2014':"Unemp_Rate2014",
                 ' Employed_2015 ':"Employed2015",
                 ' Unemployed_2015 ':"Unemployed2015",
                 'Unemployment_rate_2015':"Unemp_Rate2015",
                 ' Employed_2016 ':"Employed2016", 
                 ' Unemployed_2016 ':"Unemployed2016", 
                 'Unemployment_rate_2016':"Unemp_Rate2016",
                 'Employed_2017':"Employed2017",
                 'Unemployed_2017':"Unemployed2017",
                 'Unemployment_rate_2017':"Unemp_Rate2017",
                 'Employed_2018':"Employed2018", 
                 'Unemployed_2018':"Unemployed2018",
                 'Unemployment_rate_2018':"Unemp_Rate2018",
                 ' Employed_2019 ':"Employed2019",
                 ' Unemployed_2019 ':"Unemployed2019",
                 'Unemployment_rate_2019':"Unemp_Rate2019",
                 'Median_Household_Income_2018':"MedianHouseholdIncome2018"})
reduced_df.head()
```

```python
reduced_df["Unemployed2014"] = reduced_df["Unemployed2014"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
reduced_df["Unemployed2015"] = reduced_df["Unemployed2015"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
reduced_df["Unemployed2016"] = reduced_df["Unemployed2016"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
reduced_df["Unemployed2017"] = reduced_df["Unemployed2017"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
reduced_df["Unemployed2018"] = reduced_df["Unemployed2018"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
reduced_df["Unemployed2019"] = reduced_df["Unemployed2019"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
```

```python
reduced_df["Employed2014"] = reduced_df["Employed2014"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
reduced_df["Employed2015"] = reduced_df["Employed2015"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
reduced_df["Employed2016"] = reduced_df["Employed2016"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
reduced_df["Employed2017"] = reduced_df["Employed2017"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
reduced_df["Employed2018"] = reduced_df["Employed2018"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
reduced_df["Employed2019"] = reduced_df["Employed2019"].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '').astype(float)
```

```python
reduced_df.dtypes
```

```python
#reduced_unemploment_df to be in a specific state
ks_df = reduced_df.loc[reduced_df['stabr'] == "KS", :]
ks_df
```

```python
#determine to Kansas city
kansas_df = ks_df.loc[ks_df['AreaName'] == "Kansas", :]
kansas_df.head()
```

```python
#show columns
kansas_df.columns
```

```python
#determine (unemployed) during six years in Kansas
Kansas_reduced_df = kansas_df[['Unemployed2014', 
                               'Unemployed2015', 
                               'Unemployed2016',
                               'Unemployed2017',
                               'Unemployed2018',
                               'Unemployed2019',
                ]]
Kansas_reduced_df     
```

```python
Kansas_reduced_df = Kansas_reduced_df.rename(columns={ 
                                                     'Unemployed2014':"2014",
                                                      'Unemployed2015':"2015",
                                                      'Unemployed2016':"2016",
                                                      'Unemployed2017':"2017",
                                                      'Unemployed2018':"2018", 
                                                      'Unemployed2019':"2019"
                                                         
                                                         })
```

```python
# Collect the mean unemployeed people in  Kansas 
average_unemployed = Kansas_reduced_df.mean()

# Collect the years where data was collected
years = Kansas_reduced_df.keys()

average_unemployed
```

```python
# Plot unemployment average as a line chart
unemployed_avg, = plt.plot(years,average_unemployed , color="red", label="Unemployed Average" )

# Create a legend for our chart
plt.legend(handles=[unemployed_avg,], loc="best")
plt.title("Unemploed in Kansas(2014-2019)", fontdict={'fontweight':'bold', 'fontsize':18})
plt.xlabel("Years")
plt.ylabel("The Numbers of Unemployed")


#save image.png
#plt.savefig("Unemployment_KSRate.png", dpi=300)

# Show the chart
plt.show()
```
